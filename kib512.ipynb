#### **3.1** &emsp; Get string/unsigned char array input. As an example, we will use "abc" as user input which is equivelant to {0x61,0x62,0x63}


#### **3.2** &emsp; Append 0x90 to the end of data.

end of data delimeter is 0x90 because there was a byte delimeter needed and a way to generate this value was using the $\sigma_x$ matrix as string. $1001_2$ in bits, equal to 0x90 in hexadecimal. This delimeter is required because if data were to end in zeros, it wouldn't be possible to know where the input data ends in the matrix. For example, it would be impossible for the algorithm to differenciante between the two inputs: "abc0000" and "abc" because if padded into 8 bytes, they result in: "abc0000**0**" and "abc**00000**". But if they have a delimeter, they become "abc00001" and "abc1**0000**" which would produced a different output which results in better proper-diffusion.

### **4.0 &emsp; *Elliptic Curve Cryptography***

The Elliptic Curve of Kib512 is my customly generated Koblitz Weierstrass curve. The curve is a 64-bit curve in Galois field $\mathbb{F}_p$

#### **4.1** &emsp; tckp64k1 (t.c. kibnakamoto prime 64-bit koblitz curve with co-factor equal to 1)

$n$ is the order of the curve.

$p$ is the field size of the Galois Field $\mathbb{F}_p$, p is the largest unsigned 64-bit prime number


$p = \texttt{0xffffffffffffffc5}$

$\\ a = \texttt{0x0000000000000000}$

$\\ b = \texttt{0x0000000000000007}$

$\\ n = \texttt{0xffffffffffffffc6}$

#### **4.2 &emsp; *Elliptic Point Addition***

---
---
parameters:  $A$ (point 1), and $B$ (point 2)

$\left\{
  \begin{array}{ c l }
    \lambda = (3{A_x}^2 + a) \div (2A_y) \text{  (mod p)} & \quad \textrm{if } A_y \equiv B_y \text{ or } A_x \equiv B_x \\
    \lambda = (B_y - A_y) \div (B_x - A_x) \text{  (mod p)}                 & \quad \textrm{else}
  \end{array}
\right.$

$ R_x = \lambda^2 - A_x - B_x \text{ (mod p)}\\
  R_y = \lambda(A_x - R_x) - A_y \text{ (mod p)}
$

let division in finite field denote multiplying with modular inverse

$R$ is equal to $A + B$ on $\mathbb{F}_p$ elliptic curves

when A and B are elliptic curve points, then $A+B$ corresponds to Elliptic Point Addition
